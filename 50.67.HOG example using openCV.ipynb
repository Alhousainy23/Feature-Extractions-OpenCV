```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
import os 
import seaborn as sns 
import scipy as sc 
import pandas as pd 
os.chdir(R"D:\\AI\\AI Models\\3. OpenCV\\0- OpenCV\\Data\\Images")
filename ='Screenshot 2024-03-11 232724.png'
def main():
    #Create a HOG Descriptor 
    hog = cv.HOGDescriptor()
    #Initialize The People detector 
    hog.setSVMDetector(cv.HOGDescriptor_getDefaultPeopleDetector())
    #Load Image 
    image = cv.imread(filename)
    (bounding_boxes , weights) = hog.detectMultiScale(image,winStride=(4,4),padding=(8,8),scale=1.05)
    #Draw nounding boxes on the image
    for (x,y,w,h) in bounding_boxes:cv.rectangle(image,(x,y),(x+w,y+h),(0,0,255),4)
    cv.imshow('Image',image)
    cv.waitKey(0)
    cv.destroyAllWindows()
    plt.imshow(image)
main()
```
