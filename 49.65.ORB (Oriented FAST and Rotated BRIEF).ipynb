```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns  
import scipy as sc 
import os 
os.chdir(R"D:\\AI\\AI Models\\3. OpenCV\\0- OpenCV\\Data\\Images")
image1 = cv.imread("template.png",0)
image2 = cv.imread('Screenshot 2024-03-10 150955.png',0)
#Call orb Algorithm 
orb = cv.ORB_create()
#Used ORB --> even make detect features + descriptor between 2 images
kp1, des1 = orb.detectAndCompute(image1,None)
kp2, des2 = orb.detectAndCompute(image2,None)
image = cv.drawKeypoints(image2,kp2,None,flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
image = cv.resize(image,(600,500))
bf = cv.BFMatcher(cv.NORM_HAMMING,crossCheck=True)
#Take bf because make match between des 1, des 2
matches = bf.match(des1,des2)
#Sorted to matches 
matches = sorted(matches,key=lambda x:x.distance)
image3 = cv.drawMatches(image1,kp1,image2,kp2,matches[:15],None,flags=2)
image3 = cv.resize(image3,(600,500))
cv.imshow("Key Points", image3)
#cv.imshow('Image', image)
plt.imshow(image3)
cv.waitKey(0)
cv.destroyAllWindows()
```
