```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
import os 
os.chdir(R"D:\\AI\\AI Models\\3. OpenCV\\0- OpenCV\\Data\\Images")
image = cv.imread("blobs.jpg",0)
#Set up the simple blood detector with default parametrs 
params = cv.SimpleBlobDetector_Params() #               
#Define threshold 
#Can define thresholdStep .. See Documentation 
#The best thing will make tunning for hyperparamters for image or can used default 
params.minThreshold = 0
params.maxThreshold = 255
#Filter by area -->         
params.filterByArea = True
#params.minArea = 50 #             50
params.minArea = 40
params.maxArea = 10000 #     10000
# If you need filter to image by using color ==> filter by color (black =0)
params.filterByColor = False #Set True for cast_iron  
params.blobColor = 0 #                
#Make filter by using circularity 
params.filterByCircularity = True 
params.minCircularity =0.5 
params.maxCircularity = 1 
#filter by using Convexity 
params.filterByConvexity = True 
params.minConvexity = 0.1 
params.maxConvexity = 1 
#after detect and put parameters --> now apply 
detector = cv.SimpleBlobDetector_create(params)
#Detect blobs 
keyPoints = detector.detect(image)
print("The Number of blobs detected are : ", len(keyPoints))
#Draw Blobs 
#None --> can be replace and put another image if you need to match between 2 images
#DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS -->  this is constant and benefit from used that --> to detect a circle for each object depend on scale for eachobject
image_with_blobs = cv.drawKeypoints(image,keyPoints,None,(0,0,255),cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(image_with_blobs)
cv.imshow("KeyPoints",image_with_blobs)
cv.waitKey(0)
cv.destroyAllWindows()
#Save Result 
#cv.imwrite("Partical_Blobs_Jpg",image_with_blobs)
```
